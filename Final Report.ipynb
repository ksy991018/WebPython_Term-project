# 사용자에게 인스타그램 기반 스타벅스 메뉴 정보 제공 

## 프로젝트 목표 및 내용

- 오늘날 대학생들을 포함한 많은 사람들이 친목도모나 스터디를 목적으로 카페를 이용한다. 그 중, 제일 인기있는 프랜차이저 카페가 스타벅스이다.
- 스타벅스에는 다양한 음료, 빵, 케이크들을 판매하는데, 주문 할때 어떤 메뉴를 선택할지 고민하는 사람들이 많다. 또한, 음료나 케익의 종류가 자주 바뀌기도 한다. 
-  **사용자는 스타벅스 전체 메뉴(음료/케이크)를 한눈에 보고, 원하는 메뉴의 정보(사진,관련 해시태그)를 인스타그램 기반으로 획득할 수 소프트웨어를 개발하는 것이 목표**이다. 
- 스타벅스 메뉴 정보는 요즘 사람들이 흔히 쓰는 SNS중 하나인 인스타그램의 사진과 해시태그로 제공되고, 메뉴의 정보는 스타벅스 공식 웹사이트 정보를 기반으로 제공하고자 한다.


## 주제 선정 이유 및 이 프로젝트의 필요성

- 오늘날 카페를 찾는 사람들이 부쩍 늘어났고, 특히 스타벅스는 프랜차이저 카페들 중 **브랜드평판 1위**로 많은 사람들이 즐겨 찾는 대중적인 카페이다. 따라서, 제일 영향력있는 카페로 스타벅스가 생각되어 프로젝트 주제를 많은 카페들 중에서 스타벅스에 초점을 맞추게되었다. 
![사진](img1.jpg)  **1)사진출처(한국기업평판연구소)**


- 본인 또한 공부를 하거나 식사 후 디저트를 먹기 위해 스타벅스를 자주 찾는 편이다. 하지만, 갈 때마다 어떤 음료를 먹을지, 어떤 케이크를 먹을지 고민하는 시간이 많다. 많은 사람들이 메뉴를 정할때 흔히 말하는 결정장애(선택을 해야하는 상황에서 쉽게 결정을 내리지 못하고 우왕좌왕하는 것)를 갖고있다. 이를 해결하기 위해, 스타벅스 메뉴 관련 정보를 제공하는 소프트웨어를 생각하게 되었다. 
- 오늘날 사람들은 유행에 민감하다. 특히, 대학생들은 요즘 인기있는 음료나 케이크를 찾아 먹는 것을 좋아한다. 이를 위한 수단으로 인스타그램을 생각했다.
- 인스타그램은 대학생들이 요즘 제일 많이 사용하는 SNS로, 자신이 먹었던 맛있는 음식이나 사진을 기록할때 많이 사용한다. 이 때, 해시태그 기능으로 사진에 대한 설명을 덧붙이곤 한다. 예를들어, 스타벅스의 초코케이크를 먹었다면 **#스타벅스초코케이크#스타벅스케익#스타벅스케익추천#달달구리**과 같이 해시태그를 붙여 게시물을 업로드한다. 이렇게 게시물을 업로드하면, 돋보기 기능에서 특정 해시태그를 검색하였을 때, 그 해시태그를 포함한 게시물들과 사진들을 쭉 볼 수 있다. 
- 해시태그에 검색했을때 게시물 검색량이 많을수록 사람들이 많이 소비하고 최근 인기있는 메뉴이고, 관련 게시물에 언급된 다른 해시태그들은 그 메뉴에 대한 설명을 나타낸다고 생각된다.
- 이 소프트웨어를 통해 요즘 스타벅스의 인기있는 메뉴가 무엇인지 알아볼 수 있고, 어떤 메뉴를 선택할지 고민하는 시간을 줄일 수 있다.


## 데이터 획득 

- 필요한 데이터: **1)스타벅스 메뉴, 2)특정메뉴를 인스타그램에 검색하였을 때 나오는 게시물 검색량, 사진, 해시태그**  


1. 스타벅스 메뉴 리스트를 저장하기 위해 스타벅스 공식 웹사이트 [(링크)](https://www.starbucks.co.kr/index.do)에서 메뉴들을 웹크롤링 한다. 
    * BeautifulSoup4와 Selenium 라이브러리를 불러온 후 크롬 웹드라이버를 통해 스타벅스 음료 탭에 들어간다. 개발자 도구를 통해 모든 음료들이 menuDataSet클래스로 묶여있는것을 알 수 있고, 관련 정보들을 모두 크롤링한다. 크롤링한 데이터는 **Stabucks_menu.csv**에 저장한다.  
    

2. 인스타 [(링크)](https://www.instagram.com/)에서 돋보기 검색에, **#스타벅스+음료이름** 혹은 **#스타벅스+케이크**를 검색하여 관련 게시물에 있는 사진들과 해쉬태그들을 웹크롤링하여 저장한다.
    * 이 또한, BeautifulSoup4와 Selenium 라이브러리와 크롬 웹드라이버를 사용하여, 인스타그램 공식 사이트에 접속한다, 아이디와 비밀번호를 입력시켜 자동 로그인 시킨 후, 돋보기에 원하는 메뉴를 자동 검색한다. 검색했을 때 나오는 게시물 검색량을 저장한다. 그 후, 스크롤을 내리기 전까지 볼 수있는 이미지들을 크롤링하여 **insta_img**폴더에 해당 메뉴이름으로 저장한다. 마지막으로, 맨 처음 게시물부터 차례대로 게시물 내용에 써져있는 해시태그들을 모두 크롤링하여 **insta_tag**에 저장한다. 이 때, **driver.find_element_by_css_selector().click** 을 통해 첫번째 게시물을 클릭한 후, 해시태그 수집 후 옆에 있는 게시물로 이동하기 위해 클릭하고를 반복하여 구현한다. 만약, 검색하였을때 게시물 검색량이 0이거나 존재하지 않는 페이지면, 즉시 종료한다.

## 구현 내용 설명
  
  
1. 데이터 가공방법  
    * 스타벅스 메뉴는 menuDataSet클래스로 묶여있기 때문에, **soup.findAll('li',{'class':'menuDataSet'})** 를 통해 데이터를 크롤링한다. 그 다음, **img** 태그를 찾아 음료의 이름이 적혀있는 **alt** 속성을 가져와 csv파일에 저장한다. 처음에 코드를 작성한 후, csv파일을 열어보니 모든 음료들이 (예시: 카푸치노 음료가 있다면 카  푸  치  노로 저장) 한글자씩 저장되어있는 문제점을 발견하였다. 이에 따라, **myWriter.writerow(name.replace(" ","").strip('\n').split(','))** 를 추가하여 실행하였더니, 정상적으로 저장되었다.  
    
    * 인스타그램 돋보기에 스타벅스 메뉴를 쳤더니, 몇가지 문제점이 발견되었다. 첫번째는, 검색한 메뉴에 대한 사진이 스타벅스 메뉴가 아닌경우가 많았다. 이 프로젝트는 스타벅스 메뉴에 초점을 두고 있기 때문에, 다른 카페의 동일음료 정보를 필요로 하지않는다. 이에 따라, 스타벅스 메뉴 이름에 스타벅스가 포함되어 있지 않으면, ***스타벅스+메뉴이름*** 으로 검색하도록 구현한다. 이미 스타벅스가 메뉴이름에 포함되어있으면, 그대로 검색한다. 두번째는, 스타벅스 메뉴 이름에 "190ML", "&", "()",한자 등이 들어가는 메뉴들이 있었다. 이것들은 돋보기에 검색하기에 부적절한 이름으로 생각되어, 해당 메뉴를 검색해야할때는 **replace()** 함수로 해당 문자들을 지운 후에 검색하도록 구현한다. 세번째는, 검색한 해시태그에 관한 게시물이 0개거나 페이지 자체가 존재하지 않는 경우이다. 이런 경우에는, **try~except**을 사용하여 크롤링을 하지 않고 프로그램을 종료시키도록 구현한다.  
    * insta_tag는 크롤링한 게시물들의 해시태그들의 정보가 저장되어있는 리스트이다. 워드클라우드를 구현하기 위해서는 WordCloud의 **generate()** 함수를 사용해야하는데, 이 함수는 매개변수로 공백으로 분리된 문자열을 받는다. 따라서, **insta_tag=(" ").join(insta_tag)** 를 추가하여, 해시태그 요소들을 공백으로 분리된 하나의 문자열로 만든 후 워드클라우드를 구현한다. 
  
  
2. 소프트웨어 기능 및 동작 방법  (순서대로)

## 결론

- 이번 프로젝트의 목표는 **사용자가 스타벅스 전체 메뉴(음료/케이크)를 한눈에 보고, 원하는 메뉴의 정보(사진,관련 해시태그)를 인스타그램 기반으로 획득할 수 소프트웨어를 개발하는 것**이었다. 
- 이 소프트웨어를 통해 사용자는 스타벅스 음료나 케이크에 대한 메뉴를 한눈에 볼 수 있고, 그 중 특정 메뉴를 선택하면 인스타그램 돋보기 기능을 이용하여 관련 게시물 검색량, 사진, 해시태그에 대한 정보를 수집 가능했다. 또한, 수집된 해시태그에 대한 정보를 효과적으로 시각화하기 위해 워드클라우드를 획득하였다. 이에 따라, 프로젝트에서 추구하고자 한 기능적인 부분은 어느정도 구현하는데 성공하였다.
- 하지만, 소프트웨어를 구현하는 과정에서 몇가지 문제점들을 발견했다.
    - 1) 스타벅스 공식사이트에서 메뉴 이름들을 그대로 수집해왔는데, 수집한 이름이 인스타그램에 그대로 검색하기에는 부적절한 경우가 많았다. 특수기호, 영어, 한자가 섞여있는 메뉴들은 인스타그램에 검색하여도 페이지가 존재하지 않는 경우가 대부분이었다. 이런 경우는, replace()함수를 통해 최대한 적절한 검색어로 가공 후 진행하였지만, 여전히 검색하는데 문제가 생기는 경우가 있었다. 이런 경우, try~except문을 통해 크롤링을 하지않고 프로그램을 종료시켰지만, 더 좋은 대안이 필요할 것 같다.
    - 2) 인스타그램 특성상, 해당 해시태그를 포함하는 게시물이라고 해서 해당 해시태그에 대한 게시물이 아닌 경우도 종종 있다. 해시태그를 타고 게시물에 접속했지만, 스타벅스와 무관한 게시물인 경우라던지, 인물 사진만 올라가 있던지, 메뉴에 대한 정보를 크롤링하는데 관련없는 정보까지 모두 크롤링 하는 변수가 있었다. 이를 방지하기 위해서, 메뉴 이름에 스타벅스가 들어가있지 않으면 스타벅스+메뉴이름으로 검색하도록 구현하였다. 하지만 크롤링을 마친 후, 해시태그정보와 사진을 보면 스타벅스와 관련없는 정보들을 여전히 찾을 수 있다. 이것을 해결하기 위해서는 좀 더 엄밀하게 데이터를 가공하고 예외처리를 해줘야 할 것같다. 하지만, 인스타그램 특성상 완벽히 필요한 정보만 수집하는데는 한계가 있을것으로 보인다.
    - 3) 많은 데이터들을 크롤링하려면 꽤 긴 시간이 소요된다. 이 때문에, 게시물 탐색을 30개로 제한하였다. 시간은 비교적 짧게 걸리는 대신, 다양한 게시물의 정보를 가져오기에는 힘들었고, 해시태그 정보도 좀 더 다양성이 필요해보인다. 워드클라우드를 보면, 생각보다 해당 메뉴에 대한 구체적인 평이나 정보를 수집하기에는 한계가 있어보인다. 게시물 갯수를 좀 더 늘리고, 해시태그 뿐만 아니라 게시물 내용도 크롤링하여 워드클라우드로 구현한다면 좀 더 효과적인 정보를 제공할 것이라 본다.
    - 이러한 문제점들을 보완한다면, 좀 더 사용자에게 유익한 소프트웨어가 될 것이다.

- 이번 프로젝트를 시작하기 전에는, 크롤링을 하는 방법은 물론 크롤링이 무엇인지도 잘 몰랐다. 프로젝트 목표를 정하고, 목표하고자 한 소프트웨어를 구현하기 위해 따로 웹크롤링을 공부하고 찾아보면서, 어렵고 힘든순간도 있었고 중간중간 막히는 부분도 많았다. 하지만, 단계별로 조금씩 해내면서 배워가는 부분이 무척 많았다. 아직 미흡하고 부족한 점도 많지만, 하나의 프로젝트를 처음부터 끝까지 스스로 완성해보는 유익한 시간이었다.

#### 2) python소스코드 원본

```python
import time
import re
import sys
from selenium import webdriver
from bs4 import BeautifulSoup
import csv
from urllib.request import urlopen
from wordcloud import WordCloud
import matplotlib.pyplot as plt

driver = webdriver.Chrome('chromedriver_win32/chromedriver.exe')

with open ("Stabucks_drink.csv", 'w', encoding='utf-8',newline="") as fileWrite:
    myWriter = csv.writer(fileWrite)
    driver.get("https://www.starbucks.co.kr/menu/drink_list.do")
    time.sleep(7)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    drinks = soup.findAll('li',{'class':'menuDataSet'})
    
    #Stabucks_menu.csv파일에 스타벅스 음료 리스트 저장 
    for row in drinks:
        img_tag = row.find("img")
        name = img_tag['alt']
        myWriter.writerow(name.replace(" ","").strip('\n').split(','))

with open ("Stabucks_cake.csv", 'w', encoding='utf-8',newline="") as fileWrite:
    myWriter = csv.writer(fileWrite)
    driver.get("https://www.starbucks.co.kr/menu/food_list.do?CATE_CD=product_cake")
    time.sleep(7)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    drinks = soup.findAll('li',{'class':'menuDataSet'})
    driver.quit()
    
    #Stabucks_menu.csv파일에 스타벅스 케이크 리스트 저장 
    for row in drinks:
        img_tag = row.find("img")
        name = img_tag['alt']
        myWriter.writerow(name.replace(" ","").strip('\n').split(','))

while True:   
    print("1. 스타벅스 음료 \n2. 스타벅스 케이크")
    n = int(input("찾고싶은 정보 번호 입력 >"))
    if n==1:
        with open('Stabucks_drink.csv', 'r',encoding="utf-8") as fileRead:
            fileMatrix=[]
            for line in fileRead:
                fileMatrix.append(line.strip("\n").split(","))
        break
    elif n==2:
        with open('Stabucks_cake.csv', 'r',encoding="utf-8") as fileRead:
            fileMatrix=[]
            n=0
            for line in fileRead:
                fileMatrix.append(line.strip("\n").split(","))
        break
    else:
        print("잘못 입력하셨습니다.")
        print()

print()           
print("*** 스타벅스 메뉴목록 ***")
print()
for i in range(len(fileMatrix)):
    print(str(i+1) + "." + str(fileMatrix[i]))

print()
choice_num = int(input("원하는 메뉴 번호 입력 >"))
choice_menu = str(fileMatrix[choice_num-1]).strip("[]").replace(" ","").strip("''")

#메뉴 이름안에 "스타벅스" 들어가지 않으면 해시태그 검색할때 앞에 스타벅스 붙여주기 
if choice_menu.find("스타벅스")== -1:
    choice_menu="스타벅스"+choice_menu
#메뉴 이름에 190ML들어가면, 제거
if choice_menu.find("190ML")!=-1:
    choice_menu=choice_menu.replace("190ML","")
#메뉴 이름에 &들어가면 제거
if choice_menu.find("&")!=-1:
    choice_menu=choice_menu.replace("&","")
#메뉴 이름에 米들어가면 제거    
if choice_menu.find("米")!=-1:
    choice_menu=choice_menu.replace("米","")
#메뉴 이름에 New들어가면 제거   
if choice_menu.find("New")!=-1:
    choice_menu=choice_menu.replace("New","")
#메뉴 이름에 괄호 들어가면 제거
if choice_menu.find("(")!=-1:
    choice_menu=choice_menu.replace("(","")
if choice_menu.find(")")!=-1:
    choice_menu=choice_menu.replace(")","")

driver = webdriver.Chrome('chromedriver_win32/chromedriver.exe')
driver.get('http://www.instargram.com')
time.sleep(3)

#인스타그램 로그인
email = '아이디'  
input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
input_id.clear()
input_id.send_keys(email)

password = '' 
input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
input_pw.clear()
input_pw.send_keys(password)
input_pw.submit()
time.sleep(3)

#해시태그 검색 
url = "https://www.instagram.com/explore/tags/{}/".format(choice_menu)

driver.get(url)
time.sleep(2)

try: 
    error=0
    data = driver.find_element_by_css_selector('.g47SY').text
    print("검색 량 : " + data)
    
    #ex)1,494로 된거 쉼표제거한 후 int형태인 1494로 바꿔주기
    data_num=int(data.replace(",",""))

    #현재 directory내 insta_img 폴더 안에 검색한 해시태그 관련 사진들 저장
    #단, 검색했을때 뜨는 사진만 저장
    
    def image():
        html=driver.page_source
        soup = BeautifulSoup(html)
        insta = soup.select('.v1Nh3.kIKUG._bz0w')
        count = 0
        for i in insta:
            imgURL = i.select_one('.KL4Bh').img['src']
            with urlopen(imgURL) as f:
                count +=1
                with open('./insta_img/'+ choice_menu + str(count)+ '.jpg',"wb") as h:
                    img=f.read()
                    h.write(img)
        if count>0:
            return 1
        else:
            return 0

    if image():
        print("\n#{0} 인스타그램 사진 저장 완료!".format(choice_menu))
    else:
        print("\n#{0} 인스타그램 사진 저장 실패!".format(choice_menu))
        error=1

    #해시태그 수집

    #첫번째 게시물 클릭
    def start(driver):
        driver.find_element_by_css_selector("div._9AhH0").click()
        time.sleep(3)

    start(driver)

    #게시물의 내용중에 해시태그만 수집
    def get_tag(driver):
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        try:
            content = soup.select('div.C4VMK > span')[0].text
        except:
            content = ' '
        tags = re.findall('#[A-Za-z0-9가-힣]+', content) 
        tag = ''.join(tags).replace("#"," ") # "#" 제거 
        tag_data = tag.split() 
        return tag_data

    #화살표 모양눌러서 다음 게시물로 이동
    def move_next(driver):
        driver.find_element_by_css_selector ('a.coreSpriteRightPaginationArrow').click()
        time.sleep(3)

    #게시물이 n개 이상이면, n개 게시물의 해시태그 크롤링.
    #게시물이 n개 미만이면, 게시물 개수만큼 해시태그 크롤링
    insta_tag = []
    if data_num > 30:
        count = 30
    else:
        count = data_num

    for i in range(count-1): 
        try:
            tags = get_tag(driver)    
            for tag_one in tags: 
                insta_tag.append(tag_one)
            move_next(driver)
        except:
            time.sleep(2)
            move_next(driver)
    print()  
    print("< #{}관련 게시물 해시태그 >".format(choice_menu))
    print(insta_tag[:])
    
    insta_tag=(" ").join(insta_tag) #해시태그 요소들을 하나의 문자열로 만듬 
    wordcloud = WordCloud(background_color='white').generate(insta_tag)

    wordcloud = WordCloud(
        font_path = "C:\\Users\\KSY\\Desktop\\웹파\\프로젝트\\My_TermProject\\NanumFontSetup_TTF_BARUNPEN\\NanumBarunpenB.ttf", #폰트경로
        width = 800,
        height = 800
    )

    print("\n< #{0} 워드클라우드 >".format(choice_menu))
    wordcloud = wordcloud.generate(insta_tag)
    fig = plt.figure(figsize=(30,12))
    plt.imshow(wordcloud)
    plt.axis("off") #x,y축 숫자제거
    plt.savefig('./hashtag_wordcloud/'+ choice_menu) #hashtag_wordcloud폴더에 저장
    plt.show()
    print()
    print("\n#{0} 워드클라우드 사진 저장 완료!".format(choice_menu))
except:
    if error==0: #게시물 페이지가 존재하지 않거나 검색량이 아예 뜨지않을때
        print("죄송합니다. 검색하신 해시태그 페이지를 찾을수 없습니다.")
    elif error==1:  #게시물 검색량이 0일때 
        print("죄송합니다. 검색하신 해시태그 게시물이 없습니다.")
finally:
    driver.close()
```
